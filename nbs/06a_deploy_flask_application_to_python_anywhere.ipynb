## Introduction
This notebook shows how to deploy the flask application to `PythonAnywhere`, a free online hosting server.

## Objectives
-  Configure `PythonAnywhere`
-  Connecting to the project on `GitHub`
-  Deploy flask to `PythonAnywhere`
-  Keeping up to date from `GitHub`

## Configuration

A free accont on `PythonAnywhere` can be created via their official website:
[PythonAnywhere](https://www.pythonanywhere.com/registration/register/beginner/).



Once the sign-up is completed, the user have to verify his email, and will get a localised web link to access the service.

## Deployment

### Step 1 
Fetching the source code from `GitHub`

-  Once logged-in, navigate to the `Consoles` > `Bash`.





-  Type the following commands to create an SSH-Key for `GitHub`.

```python
# Generate an ssh-key
$ ssh-keygen
# Read the generated ssh-key
$ cat ~/.ssh/id_rsa.pub
```


-  Copy the ssh-key generated and add to your `GitHub` account by navigating to `Settings` > `SSH and GPG keys` > `New SSH key`.



-  Clone the project file from `GitHub`.

```python
$ git clone git@github.com:khhaledahmaad/attendance_monitoring_via_sms.git
```



### Step 2
Setting up a `virtualenv` for `Pythonanywhere` server

-  Create a `virtualenv` for your web app.
```python
$ mkvirtualenv .twilio --python=/usr/bin/python3.10
```



-  Change to your cloned project directory.
```python
$ cd attendance_monitoring_via_sms
```



-  Install all dependencies from `requirements.txt` file.
```python
$ pip install requirements.txt
```



## Step 3 
Configuring the web

-  Navigate to `Web` > `Add new web app`, then select `Manual configuration` > `Flask` > `Python 3.10`, the latest version for the application. `PythonAnywhere` will creat a publicly accessible link for the app. For more details, follow: https://help.pythonanywhere.com/pages/Flask.







-  Navigate to `Web` then select your newly created app.



-  Scroll down to `Code` and `Virtualenv`.



-  Enter the location of your cloned project directory to `Enter the path to your web app source code` and to the `Working directory` and the installed virtualvenv to `Enter path to a virtualenv, if desired`.



-  Naviagete to `WSGI configuration file` from `Code`, delete or comment out everything apart from the code in the `Flask` section shown in the picture below, and edit the `import stetement` as per the name of the the cloned project's main `Flask` script file (app.py) and save it. Also, edit the `path` to your cloned project directory.




-  Navigate to `Security` and enable `Force HTTPS`.



-  Finally navigate to `Web` and press `Reload` link for your application and access the application running on the link under `Configuration for`.



- The `Flask` application will run, and you can navigate to the endpoints.

__N.B.:__ As we are running a free version of `Pythonanywhere`, it is possible to see some limitations as to running the application locally, such as accessing endpoints related to sending emails. This is due to the limited capacity and memory usage allocated for the free account. An upgraded account on `Pythonanywhere` is expected to demonstrate much better speed and performance.



## Keeping Up to Date

Make sure to update the application everytime the github repo is updated by using the following commands in the `Pythonanywhere` `Bash` console:

```python
$ cd attendance_monitoring_via_sms

$ git pull
```


Once the application is deployed, it is really easy to update any changes directly from `GitHub` following the steps mentined above.
