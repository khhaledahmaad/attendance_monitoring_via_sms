## Introduction
This notebook shows how to deploy the flask application to `PythonAnywhere`, a free online hosting server.

## Objectives
-  Configure `PythonAnywhere`
-  Deploy flask to `PythonAnywhere`

## Configuration

A free accont on `PythonAnywhere` can be created via their official website:
[PythonAnywhere](https://www.pythonanywhere.com/registration/register/beginner/)



Once the sign-up is completed, the user have to verify his email, and will get a localised web link to access the service.

## Deployment

### Step 1 

-  Once logged-in, the user need to navigate to `Web` > `Add new web app`, then select `Flask` and the preferred version for the application. `PythonAnywhere` will creat a publicly accessible link for the app. For more details, follow: https://help.pythonanywhere.com/pages/Flask



### Step 2

-  To upload the files for the `Flask` application, navigate to `Files` and create a directory for your application.



-  Navigate to the directory created, and create and upload all the folders and files for the `Flask` applications, including the `virtualenv` folder (`.twilio`).



### Step 3

-  Navigate to `Web` then select your newly created app.



-  Scroll down to `Code` and `Virtualenv`.



-  Naviagete to `WSGI configuration file` from `Code` and edit the `import stetement` as per the name of the main `Flask` script file (app.py) and save it.



-  Select `Enter path to virtualenv, if desired` and enter and save the location to your uploaed `virtualenv` directory (`.twilio`) within the `Flask` application directory created from `Step 2`.



-  Finally navigate to `Web` and press `Reload` link for your application and access the application running on the link under `Configuration for`.
